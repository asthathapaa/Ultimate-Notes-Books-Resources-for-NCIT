```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([2,3,5,7])

slope = 0
intercept = 0
learning_rate = 0.01
cost_value = []
for i in range(100):
  y_pred = slope * x + intercept
  error = y - y_pred
  cost = np.mean(error**2)
  print(f"Slope:{slope:.3f}, Intercept: {intercept:.3f}, Cost:{cost:.3f}")
  cost_value.append(cost)
  partial_slope = -2/len(x)*np.sum(x*(error))
  partial_intercept = -2/len(x)*np.sum(error)
  slope = slope - learning_rate*partial_slope
  intercept = intercept - learning_rate*partial_intercept
print(f"Final slope:{slope:.3f} \nFinal Intercept: {intercept:.3f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(cost_value)
plt.show()
```
