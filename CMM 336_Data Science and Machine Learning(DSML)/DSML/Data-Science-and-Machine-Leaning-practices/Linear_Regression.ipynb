```python
import pandas as pd

data={
    'X' :[1,2,3,4,5],
    'Y' : [1,2,1.3,3.75,2.25]
}
df=pd.DataFrame(data)
print(df)
```

```python
mean_x = df['X'].mean()
print(mean_x)
```

```python
mean_y = df['Y'].mean()
print(mean_y)
```

```python
variance_x = df['X'].var()
print(variance_x)
```

```python
cov = df.cov()
print(cov)
```

```python
slope = cov.iloc[0,1]/variance_x
print(slope)
```

```python
a = mean_y- mean_x*slope
print(a)
```

```python
y = a + slope*df['X']
print(y)
```

```python
cov_xy= df['X'].cov(df['Y'])
print(cov_xy)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['X'],df['Y'])
plt.plot(df['X'],y,color='red')
plt.show()
```

```python
print('The simple linear regression equation is: ')
print(f'Y = {a} + {slope}X')
```

```python
#prediction for the new unseen data
y_n=slope*6+a
print("The new predicted value for 6th item :",y_n)
```
