```python
import pandas as pd
import numpy as np

data ={
    'X1': [1,2,3,4],
    'X2': [4,5,8,2],
    'Y': [1,6,8,12],
    'Z':[1,1,1,1],
}
df=pd.DataFrame(data)
print(df)
```

```python
print("Original DataFrame:")
print(df)

# Extract matrix (as a NumPy array)
matrix1 = df[['Z','X1', 'X2']].values


# Print the matrix
print("\nMatrix:")
print(matrix1)
```

```python
matrix2 = df[['Y']].values
print(matrix2)
```

```python
np.transpose
```

```python
transposed_matrix1 = matrix1.T

# Print the transposed matrix
print(transposed_matrix1)
```

```python
XTXinv = np.dot(transposed_matrix1, matrix1)  # or matrix1 @ matrix2

# Print the result
print("Matrix multiplication result:")
print(XTXinv)
```

```python
from sklearn.linear_model import LinearRegression
x = df[['X1','X2']] # Corrected indentation, use [[]] for multiple columns
y = df['Y']
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
import numpy as np

# Actual values (y) and predicted values (y_hat)
y_actual = np.array([1, 6, 8, 12])
y_predicted = np.array([1.5689, 5.0282, 8.3498, 12.1574])

# Calculate Mean Squared Error
mse = np.mean((y_actual - y_predicted) ** 2)

print(f"Mean Squared Error: {mse}")
```

```python
slope =model.coef_
print(slope)
```

```python
intercept = model.intercept_
print(intercept)
```

```python
print(f"Multiple linear equation : y= {intercept:.2f} + {slope[0]:.2f}X1+ {slope[1]:.2f}X2")
```

```python
import matplotlib.pyplot as plt

# Actual values and predicted values
y_actual = [1, 6, 8, 12]
y_predicted = [1, 5, 9, 11]

# Plot actual vs predicted values
plt.plot(y_actual, label='Actual', marker='o')
plt.plot(y_predicted, label='Predicted', marker='x')

# Add labels and legend
plt.title('Actual vs Predicted')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Scatter plot of actual vs predicted
plt.scatter(y_actual, y_predicted, color='blue', label='Predicted vs Actual')
plt.plot([min(y_actual), max(y_actual)], [min(y_actual), max(y_actual)], color='red', linestyle='--', label='y=x (perfect prediction)')

# Add labels and legend
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.show()

```
